```python
import pandas as pd
```

```python
df=pd.read_csv("SMSSpamCollection.txt",sep='\t',names=["label","message"])
```

```python
df.head()
```

### Data cleaning an preprocessing

```python
import nltk
import re
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df["message"][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

#### Bag of words  : COunt Vectorizer

```python
from sklearn.feature_extraction.text import  CountVectorizer
```

```python
cv=CountVectorizer(max_features=2500)
```

```python
cv.fit_transform(corpus).toarray()
```

```python
binary_bow_cv=CountVectorizer(max_features=2500,binary=True)
```

```python
binary_bow_cv.fit_transform(corpus).toarray()
```

```python
binary_bow_cv.vocabulary_
```

### Ngram

```python
n_gram_cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(1,1))
X=n_gram_cv.fit_transform(corpus)
```

```python
n_gram_cv.vocabulary_
```

```python
n_gram_cv=CountVectorizer(max_features=1000,binary=True,ngram_range=(1,2))
X=n_gram_cv.fit_transform(corpus)
n_gram_cv.vocabulary_
```
